### Re-initializing `quikr_car` and re-applying cleaning steps

It appears the `quikr_car` DataFrame was accidentally replaced by just the `kms_driven` Series during the previous cleaning process. To fix this and ensure the 'kms_driven' column is correctly converted within the DataFrame, we will re-initialize `quikr_car` from your `backup` and then re-run the cleaning steps for 'year', 'Price', and 'kms_driven'.

```python
# Re-initialize car from the backup
car = backup.copy()
```

#### Cleaning 'year' column

```python
# Filter out non-numeric entries in 'year' and convert to int
car = car[car['year'].astype(str).str.isnumeric()]
car['year'] = car['year'].astype(int)
```

#### Cleaning 'Price' column

```python
# Filter out 'Ask For Price' entries and convert to int
car = car[car['Price'] != "Ask For Price"]
car['Price'] = car['Price'].str.replace(',', '').astype(int)
```

#### Cleaning 'kms_driven' column and converting to int

```python
# Extract numeric part, remove commas, and filter non-numeric entries
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
car = car[car['kms_driven'].str.isnumeric()]
# Convert 'kms_driven' column to integer type
car['kms_driven'] = car['kms_driven'].astype(int)
```

Now, let's check the information of the `quikr_car` DataFrame to confirm the changes.

```python
car[car['fuel_type'].isna()]
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
car = car.reset_index(drop = True)
```

```python
car.describe()
```

```python
car[car['Price']>6e6]
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

```python
car.to_csv('Cleaned_car_data.csv')
```

### Model

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
```

```python
print(sklearn.__version__)
```

```python
Y = car['Price']
X = car.drop('Price', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder= 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred = pipe.predict(X_test)
print(r2_score(Y_test,y_pred))
```

```python
scores=[]
for i in range (1000):
  X_train, X_test, Y_train, Y_test = train_test_split ( X,Y, test_size = 0.2, random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(X_train, Y_train)
  y_pred = pipe.predict(X_test)
  scores.append(r2_score(Y_test, y_pred))
```

```python
print(np.argmax(scores))
print(np.max(scores))
print(scores[np.argmax(scores)])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split ( X,Y, test_size = 0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(X_train, Y_train)
y_pred = pipe.predict(X_test)
r2_score(Y_test, y_pred)
```
